````python
#1) Seu código aqui

def consultarDados(federal:str,cidade:str):
    url = f"http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={federal.upper()}&tipo_interessado={cidade.capitalize()}"
    return pd.DataFrame(requests.get(url).json()['items'])
````

````python
# 2) Seu código aqui

df2 = consultarDados('MG', 'Estado')
df2["status"].value_counts()["Arquivado por decurso de prazo"]
````
